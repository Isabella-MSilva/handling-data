```python
def corrigir_nomes(corrigir):
    corrigir = corrigir.replace('Nutri?’o', 'Nutrição').replace('Engenharia MecŸnica', 'Engenharia Mecânica').replace('Ci?ncias Biol½gicas', 'Ciências Biológicas').replace('Letras - Ingl?s', 'Letras - Inglês').replace('Educa?’o F­sica', 'Educação Física').replace('Educa?’o do Campo	', 'Educação do Campo').replace('Servi?o Social', 'Serviço Social').replace('T?cnico em Informÿtica', 'Técnica em informática').replace('T?cnico em Anÿlises Cl­nicas', 'Técnico em Análises Clínicas').replace('T?cnico em Enfermagem', 'Técnico em Enfermagem').replace('T?cnico em Radiologia', 'Técnico em Radiologia').replace('T?cnico em Saœde Bucal', 'Técnico em Saúde Bucal').replace('Mestranda - P½s Gradua?’o em Ci?ncias da Saœde', 'Mestranda - Pós Graduação em Ciências da Saúde').replace('T?cnico em Farmÿcia', 'Técnico em Farmácia').replace('Licenciatura em Educa?’o do Campo', 'Licenciatura em Educação do Campo').replace('?', 'ç').replace('Educa?’o do Campo	', 'Educação do Campo')
    return corrigir


importa_df['Curso'] = importa_df['Curso'].apply(corrigir_nomes)
importa_df.rename(columns={'Per­odo de Recebimento da Bolsa': 'Período de Recebimento da Bolsa', 'Institui?’o/…rg’o Responsÿvel Pelo Pagamento': 'Instituição responsável pelo pagamento' })
importa_df.head()
```
